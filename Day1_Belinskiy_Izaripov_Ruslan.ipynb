### **Container types**
list - *list()*
* ['a','a','b','c'] > can be appended with any other element
> *ex : ['a','a','b','c'].append('d') will return ['a','a','b','c','d']*
* each element of the list is associated to an index. The first element of the list has a 0 index, the second has a 1 index etc...
> *ex : in list_ex =['a','a','b','c'] the first element is 'a' and its index is 0. Thus, list_ex[0] will return 'a', list_ex[2] will return 'b' etc...

tuple - *tuple()*
* (a,b,c) > immuable, you cannot add any element to a tuple

set - *set()*
* {'a','b','c'} > set of unique values of a list
> *ex : set(['a','a','b','c']) will return {'a','b','c'}*

dictionnary - *dict()*
* {'a':'apple','b':'banana','c':'carrot'} > key/value association storage that you can call. Each key is also associated to an index
> *ex : dict_ex = {'a':'apple','b':'banana','c':'carrot'} \ dict_ex['a'] will return 'apple' and dict_ex[0] will also return 'apple'*


### **variable assignement**
any base or container type can be "stored" into a variable for easy calling. the variable name must follow some rules:
* It can be only one word 
> *my variable = 'apple'* is not a good variable name
> *myvariable = 'apple'* is a good variable name
* It can use **only** letters, numbers, and the underscore (_) character
* It can’t begin with a number

####Exercice 9
create a function called dict_squared(integer) that returns a dictionnary where the keys are the integers smaller than your number and the values are the squared integers. For example, dict_squared(3) will return {1:1,2:4,3:9}

````python
def dict_squared(integer):
  dict = {}
  i = 0
  while i < integer+1:
    dict[i].append(i**2)
    i = i+1
````

````python
dict_squared(3)
````

####Exercice 10
Create a function called even_exp(number) that returns number/2 if the number is even and number*2 if the number is odd

hint : to check if a number is even or odd, you can use the % sign https://www.freecodecamp.org/news/the-python-modulo-operator-what-does-the-symbol-mean-in-python-solved/

````python
def even_exp(number):
  if number % 2 == 0:
    return number / 2
  else :
    return number * 2
````
